```python
cat_var = ['grade','home_ownership','verification_status','purpose']
```

```python
for var in cat_var:
    cat_list = 'var'+var
    cat_list = pd.get_dummies(loan[var],prefix=var)
    df1= loan.join(cat_list)
    loan = df1
```

```python
loan.sample()
```

```python
loan.shape
```

```python
loan.drop(columns=cat_var,inplace=True)
```

```python
loan.columns.to_list()
```

```python
x= loan.drop(columns='loan_status')
```

```python
y= loan['loan_status']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
log_reg = LogisticRegression(max_iter=1000)
```

```python
log_reg.fit(X_train,Y_train)
```

```python
y_pred = log_reg.predict(X_test)
y_pred
```

```python
print("Accuray is ",log_reg.score(X_train,Y_train))
```

```python
print("Accuray is ",log_reg.score(X_test,Y_test))
```
